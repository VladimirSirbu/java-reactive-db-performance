### 1. Response Time - ms (latency)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the file
file_path = 'results/results_2024-06-17_15:27:07.log'
df = pd.read_csv(file_path, sep=',')

# Filter the DataFrame to include only specific descriptions
df_filtered = df[df['description'].isin(
    [
        'WebFlux R2DBC',
        'Web MVC JDBC',
        'Web MVC R2DBC',
        'WebFlux JDBC',
        'Quarkus_R2DBC',
        'VertX_PG_client'
    ])][['description', 'lat_avg']]

# Group by description and compute the mean of lat_avg
grouped_data = df_filtered.groupby(['description'])['lat_avg'].mean().reset_index()

# grouped_data.to_csv('output1.csv', index=False)

# Plot setup
plt.figure(figsize=(12, 8))

# Colors for each description
colors = {
    'WebFlux R2DBC': 'red',
    'Web MVC JDBC': 'blue',
    'Web MVC R2DBC': 'green',
    'WebFlux JDBC': 'orange',
    'Quarkus_R2DBC': 'cornflowerblue',
    'VertX_PG_client': 'pink'
}

# Create a bar chart
plt.bar(grouped_data['description'], grouped_data['lat_avg'], color=[colors[desc] for desc in grouped_data['description']])

# Customize the plot
plt.title('Average Latency')
plt.xlabel('Description')
plt.ylabel('Average Latency (ms)')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.show()
```
